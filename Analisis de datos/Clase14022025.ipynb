Reemplazar un elemento

```python
frutas = ["manzana", "banana", "cereza", "limon", "naranja"]
frutas [3]= "pera"
print (frutas)
```

Agregar un elemento

```python
frutas = ["manzana", "banana", "cereza", "limon", "naranja"]
print (frutas)
frutas.append("uva") #Agrega un elemento a la lista
print (frutas)
```

Insertar un elemento

```python
frutas = ["manzana", "banana", "cereza", "limon", "naranja"]
frutas.insert (2, "mandarina")#inserta un nuevo elemento en la posición
print(frutas)
frutas.insert (4, "uva")
print(frutas)
```

Remover un elemento

```python
frutas = ["manzana", "banana", "cereza", "limon", "naranja"]
frutas.remove("banana")#Remueve el elemento que coincide
print(frutas)
```

Eliminar el ultimo elemento y guardarlo en una variable

```python
frutas = ["manzana", "banana", "cereza", "limon", "naranja"]
print(frutas)
ultimo = frutas.pop()#elimina la ultima posición y la asigna a la variable
print(frutas)
print(ultimo)
```

Ciclo For o Para

```python
frutas = ["manzana", "banana", "cereza", "limon", "naranja"]
for frutas in frutas:
  print("La fruta es:",frutas)
```
