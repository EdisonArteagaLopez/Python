```r
# PASO 1: Configurar el entorno
# Instalar y cargar paquetes necesarios
install.packages("tidyverse")
library(tidyverse)

# PASO 2: Importar datos
# Desde un archivo CSV local (primero subir el archivo a Colab)
datos <- read_csv("datos.csv")

# Desde una URL
datos <- read_csv("https://raw.githubusercontent.com/plotly/datasets/master/iris.csv")

# Ver estructura del dataframe
glimpse(datos)

# PASO 3: Exploración básica de datos
# Primeras filas
head(datos, 5)

# Resumen estadístico
summary(datos)

# Dimensiones
dim(datos)

# Nombres de columnas
colnames(datos)

# PASO 4: Limpieza de datos
# Eliminar filas con NA
datos_limpios <- na.omit(datos)

# O rellenar NA con la media
datos_limpios <- datos %>%
  mutate(Sepal.Length = replace_na(Sepal.Length, mean(Sepal.Length, na.rm = TRUE)))

# Renombrar columnas
datos <- datos %>%
  rename(especie = Species)

# Convertir tipos de datos
datos$especie <- as.factor(datos$especie)

# Filtrar datos
setosa <- datos %>%
  filter(especie == "setosa")

# PASO 5: Manipulación de dataframes
# Seleccionar columnas
datos_seleccion <- select(datos, Sepal.Length, Sepal.Width, especie)

# Crear nueva columna
datos <- datos %>%
  mutate(ratio = Sepal.Length / Sepal.Width)

# Agrupar y resumir
datos %>%
  group_by(especie) %>%
  summarize(
    media_longitud = mean(Sepal.Length),
    max_ancho = max(Sepal.Width)
  )

# PASO 6: Visualizaciones básicas
# Histograma
ggplot(datos, aes(x = Sepal.Length)) +
  geom_histogram(bins = 20, fill = "blue", color = "black") +
  ggtitle("Distribución de Longitud del Sépalo")

# Diagrama de dispersión
ggplot(datos, aes(x = Sepal.Length, y = Sepal.Width, color = especie)) +
  geom_point(size = 3) +
  labs(title = "Relación Longitud-Ancho del Sépalo")

# Boxplot
ggplot(datos, aes(x = especie, y = Sepal.Length, fill = especie)) +
  geom_boxplot() +
  theme_minimal()

# PASO 7: Flujo completo de análisis
# 1. Importar datos
# 2. Exploración inicial
# 3. Limpieza y preprocesamiento
# 4. Análisis estadístico
# 5. Visualización
# 6. Conclusiones

# Ejemplo completo con datos de prueba:
data(mtcars)
mtcars %>%
  mutate(cyl = as.factor(cyl)) %>%
  filter(hp > 100) %>%
  group_by(cyl) %>%
  summarize(avg_mpg = mean(mpg)) %>%
  ggplot(aes(x = cyl, y = avg_mpg)) +
  geom_col(fill = "darkgreen") +
  labs(title = "Consumo promedio por cilindrada")
```
