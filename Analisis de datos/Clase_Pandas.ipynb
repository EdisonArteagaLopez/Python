```python
#Histograma
edades_simuladas = np.random.normal(loc=30, scale=10, size=1000)
plt.hist(edades_simuladas, bins=20, edgecolor='black')
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
titanic_df.hist(figsize=(12, 8))
plt.show()
```

```python
#selección de columnas
nombres_edades = titanic_df[['Name', 'Age']]
print(nombres_edades.head())
```

```python
#Fltrado: pasejeros mayores de 60 años
titanic_df[titanic_df['Age'] > 60]
```

```python
# Valor nulo en la edad
titanic_df['Age'].isnull().sum()
```

```python
#rellenar valores nulos con la mediana
mediana_edad = titanic_df['Age'].median()
titanic_df['Age']=titanic_df['Age'].fillna(mediana_edad)

#Agrupar por sexo y calular la tase de supervivencia
sobrevivencia_por_sexo = titanic_df.groupby('Sex')['Survived'].mean()
print(sobrevivencia_por_sexo)
```

```python
#conteo por clase de pasajero

titanic_df['Pclass'].value_counts()
```

```python
#Medidas de tendecia central y dispersión

media_edad = titanic_df['Age'].mean()
mediana_edad = titanic_df['Age'].median()
moda_edad = titanic_df['Age'].mode()[0]
desviacion_estandar_edad = titanic_df['Age'].std()

print("Media de Edad:", media_edad)
print("Mediana de Edad:", mediana_edad)
print("Moda de Edad:", moda_edad)
print("Desviación Estándar de Edad:", desviacion_estandar_edad)
```

```python
# Calculo de max y min
max_edad = titanic_df['Age'].max()
min_edad = titanic_df['Age'].min()

print("Edad máxima:", max_edad)
print("Edad mínima:", min_edad)
```

```python
# Calculo max y min Fare
max_fare = titanic_df['Fare'].max()
min_fare = titanic_df['Fare'].min()

print("Tarifa máxima:", max_fare)
print("Tarifa mínima:", min_fare)

#Calculos
print("Suma total de tarifas:", titanic_df['Fare'].sum())
print("Promedio de tarifas:", titanic_df['Fare'].mean())
print("Tarifa más alta:", titanic_df['Fare'].max())
print("Tarifa más baja:", titanic_df['Fare'].min())
```

```python
# Suma fare por psclass
suma_fare_por_pclass = titanic_df.groupby('Pclass')['Fare'].sum()
print(suma_fare_por_pclass)
```

```python
#Visualización
plt.figure(figsize=(8, 6))
plt.scatter(titanic_df['Age'], titanic_df['Fare'], alpha=0.5)
plt.title('Relación entre Edad y Tarifa')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
```

```python
#Grafico de barras
sobrevivientes_por_clase = titanic_df.groupby('Pclass')['Survived'].sum()
plt.figure(figsize=(8, 6))
sobrevivientes_por_clase.plot(kind='bar', color='skyblue')
plt.title('Número de Sobrevivientes por Clase')
plt.xlabel('Clase')
plt.ylabel('Número de Sobrevivientes')
plt.xticks(rotation=0)
plt.show()
```

```python
# Histograma por edades
plt.figure(figsize=(8, 6))
plt.hist(titanic_df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
```
