### (1) LIBRERÍAS A UTILIZAR ###

```python
#Se importan la librerias a utilizar
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
print ("Las librerias han sido cargadas")
```

### (2) Carga de los Datasets (archivos de datos ###

```python
#Se importa los datos a utilizar de la web
#url_test = 'https://storage.googleapis.com/kaggle-competitions-data/kaggle/3136/test.csv'
#url_train = 'https://storage.googleapis.com/kaggle-competitions-data/kaggle/3136/train.csv'
url_test = '/content/drive/MyDrive/Datasets/Titanic/test.csv'
url_train = '/content/drive/MyDrive/Datasets/Titanic/train.csv'
###############################################
###   Creamos los dataframes respectivos    ###
###############################################
df_test = pd.read_csv(url_test)
df_train = pd.read_csv(url_train)
print("===>>> veamos las primeras filas de test")
print(df_test.head())
print("===>>> veamos las primeras filas de train")
print(df_train.head())
```

### (3) COMPRENSION DE LA DATA ###

```python
#Verifico la cantidad de datos que hay en los dataset
print("--------------------------")
print('Cantidad de datos:')
print("train : ",df_train.shape)
print("test : ",df_test.shape)

#Verifico el tipo de datos contenida en ambos dataset
print("--------------------------")
print('Tipos de datos:')
print("train : ",df_train.info())
print("test : ",df_test.info())
print("--------------------------")
#Verifico los datos faltantes de los dataset
print('Datos faltantes:')
print("train : ",pd.isnull(df_train).sum())
print("test : ",pd.isnull(df_test).sum())
print("--------------------------")
#Verifico las estadísticas del dataset
print('Estadísticas del dataset:')
print("train : ",df_train.describe())
print("test : ",df_test.describe())
```

###  (4) PREPROCESAMIENTO DE LA DATA ###


```python
#Cambio los datos de sexos en números
print("Cambio los datos de sexos en números")
print(df_train['Sex'].unique)
df_train['Sex'].replace(['female','male'],[0,1],inplace=True)
df_test['Sex'].replace(['female','male'],[0,1],inplace=True)
```

```python
#Cambio los datos de embarque en números
print("Cambio los datos de embarque en números")
df_train['Embarked'].unique
df_train['Embarked'].replace(['Q','S', 'C'],[0,1,2],inplace=True)
df_test['Embarked'].replace(['Q','S', 'C'],[0,1,2],inplace=True)
```

```python
#Reemplazo los datos faltantes en la edad por la media de esta columna
print("Reemplazo los datos faltantes en la edad por la media de esta columna")
print("media de la edad en train : ",df_train["Age"].mean())
print("media de la edad en ntest : ",df_test["Age"].mean())
promedio = 30
df_train['Age'] = df_train['Age'].replace(np.nan, promedio)
df_test['Age'] = df_test['Age'].replace(np.nan, promedio)
print("Edad cambiada")
print(df_train['Age'])
print(df_test['Age'])
```

```python
#Creo varios grupos de acuerdo a bandas de las edades
print("Creo varios grupos de acuerdo a bandas de las edades")
#Bandas: 0-8, 9-15, 16-18, 19-25, 26-40, 41-60, 61-100
bins = [0, 8, 15, 18, 25, 40, 60, 100]
names = ['1', '2', '3', '4', '5', '6', '7']
df_train['Age'] = pd.cut(df_train['Age'], bins, labels = names)
df_test['Age'] = pd.cut(df_test['Age'], bins, labels = names)
print("bandas : df_train")
print(df_train)
print("bandas : df_test")
print(df_test)
```

```python
#Se elimina la columna de "Cabin" ya que tiene muchos datos perdidos
print("Se elimina la columna de \"Cabin\" ya que tiene muchos datos perdidos")
df_train.drop(['Cabin'], axis = 1, inplace=True)
df_test.drop(['Cabin'], axis = 1, inplace=True)
```

```python
#Elimino las columnas que considero que no son necesarias para el analisis
print("Elimino las columnas que considero que no son necesarias para el analisis")
df_train = df_train.drop(['PassengerId','Name','Ticket'], axis=1)
df_test = df_test.drop(['Name','Ticket'], axis=1)
print("train : ")
df_train.describe()
print("test : ")
df_test.describe()
```

```python
#Se elimina las filas con los datos perdidos
print("Se elimina las filas con los datos perdidos")
df_train.dropna(axis=0, how='any', inplace=True)
df_test.dropna(axis=0, how='any', inplace=True)
```

```python
#Verifico los datos
print("Cantidad de nulos")
print(pd.isnull(df_train).sum())
print(pd.isnull(df_test).sum())
print("Info de los dataframes")
print(df_train.shape)
print(df_test.shape)
print("primeras filas de los dataframes")
print(df_test.head())
print(df_train.head())
```

### (5) APLICACIÓN DE ALGORITMOS DE MACHINE LEARNING ###

```python
#####################################################################
# Para el entrenamiento ...
# Separo la columna con la información de los sobrevivientes
X = np.array(df_train.drop(['Survived'], axis=1))
y = np.array(df_train['Survived'])
#####################################################################
## !!!!!  Ojo aqui !!!!
#####################################################################
# Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

## Regresión logística
## Aqui hace el entrenamiento (fit) del algoritmo
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
##_Realiza la prediccion de prueba
Y_pred = logreg.predict(X_test)
print('Precisión Regresión Logística:')
print(logreg.score(X_train, y_train))

##Support Vector Machines
svc = SVC()
svc.fit(X_train, y_train)
Y_pred = svc.predict(X_test)
print('Precisión Soporte de Vectores:')
print(svc.score(X_train, y_train))

##K neighbors
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, y_train)
Y_pred = knn.predict(X_test)
print('Precisión Vecinos más Cercanos:')
print(knn.score(X_train, y_train))
```

### (6) PREDICCIÓN UTILIZANDO LOS MODELOS ###

```python
ids = df_test['PassengerId']
df_predic = df_test.drop(['PassengerId'], axis=1)
print("tipos de datos predicción")
print(df_predic.dtypes)

# Convert object columns in df_predic to numerical using one-hot encoding
#df_predic = pd.get_dummies(df_predic, columns=df_predic.select_dtypes(include=['object']).columns)

# Convert object columns in df_predic to numerical using one-hot encoding
# Convert all object type columns to numerical using one-hot encoding
for col in df_predic.select_dtypes(include=['object']).columns:
    df_predic = pd.concat([df_predic, pd.get_dummies(df_predic[col], prefix=col)], axis=1)
    df_predic.drop(col, axis=1, inplace=True)

# Ensure all columns are numeric
df_predic = df_predic.astype(float)

###Regresión logística
print('Predicción Regresión Logística:')
prediccion_logreg = logreg.predict(df_predic)
out_logreg = pd.DataFrame({ 'PassengerId' : ids, 'Survived': prediccion_logreg })
print(out_logreg.head())

##Support Vector Machines
print('Predicción Soporte de Vectores:')
prediccion_svc = svc.predict(df_test.drop('PassengerId', axis=1))
out_svc = pd.DataFrame({ 'PassengerId' : ids, 'Survived': prediccion_svc })
print(out_svc.head())

##K neighbors
print('Predicción Vecinos más Cercanos:')
prediccion_knn = knn.predict(df_test.drop('PassengerId', axis=1))
out_knn = pd.DataFrame({ 'PassengerId' : ids, 'Survived': prediccion_knn })
print(out_knn.head())
```
