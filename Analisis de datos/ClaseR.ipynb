```r
#ejercicio 3

es_primo <- function (n) {
  if (n<2) return(FALSE)
  for (i in 2:sqrt(n)){
    if (n %% i ==0) return (FALSE)
  }
  return (TRUE)
}

primos <- c()
num<-2

while (length(primos)< 10){
  if(es_primo(num)){
    primos <- c(primos, num)
  }
  num <- num+1
}
cat ("Los primeros 10 numeros primos son: ", primos, "\n")
```

```r
#ejercicio 4

#Divisores de un numero

numero <- 40
divisores <- c()

for (i in 1:numero){
  if (numero %% i==0){
    divisores <- c(divisores, i)
  }
}
cat("Los divisores  de ", numero, "son = ", divisores, "\n")
```
