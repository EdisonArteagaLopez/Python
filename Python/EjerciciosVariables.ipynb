Solicite al usuario el nombre y apellido, peso, estatura. Donde trabaja, horas al día que trabaja y cuanto le pagan por hora.

Al final me entregue el nombre y apellido del usuario en mayusculas, calcule el imc y me diga cuanto ganaria si trabaja 30 días al mes.


```python
nombre= input("Escribe tu nombre: ")
apellido = input("Escribe tu apellido: ")
peso= float(input("Digita tu peso en kg: "))
estatura = float (input("Digita tu estatura en metros: "))
lugar_de_trabajo = (input("Escribe el lugar donde trabajas: "))
horas_por_dia= float (input("Cuantas horas trabajas al día: "))
pago_hora= float (input("¿Cuanto te pagan por hora?"))
nombre_completo= f"{nombre.upper()} {apellido.upper()}"
print(f"Nombre completo en mayuscula: {nombre_completo}")
imc = peso/(estatura)**2
print (f"Tu IMC es:  {imc:.2f}")
pago_mensual= horas_por_dia * pago_hora * 30
print (f"el pago mensual sería:  {pago_mensual:.2f} pesos")

```

```python
# Condicionales
print ("Evaluar si un numero es mayor que 10 ")
num1 = int (input ("ingresa un numero entero "))
if num1 > 10:
  print(f"El numero {num1} es mayor que 10 ")
elif num1 == 10:
  print(f"El número {num1} es igual a 10 ")
else:
  print(f"EL número {num1} es menor que 10 ")
```
