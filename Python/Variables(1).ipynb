#Ejemplo listas

```python
dias_de_la_semana = ["lunes", "martes", "miercoles", "jueves", "viernes", "sabado", "domingo"]
print (dias_de_la_semana)
```

```python
# Imprimir d√≠a miercoles

print (dias_de_la_semana[2])
```

```python
#Reemplazar un elemento

dias_de_la_semana [3] = "Daniela"
print (dias_de_la_semana)
```

```python
# Sumar un elemento a la lista
dias_de_la_semana.append("Angie")
print(dias_de_la_semana)
```

```python
#Borrar un elemento
dias_de_la_semana.remove("Daniela")
print(dias_de_la_semana)
```

```python
#Imprimir el ultimo elemento de la lista
print(dias_de_la_semana[6])
print(dias_de_la_semana[-2])
```
