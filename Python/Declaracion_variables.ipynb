#Declaración de listas

```python
# lista_de_nombre = [0, 1, 2, 3, 4]
lista_de_nombres= ["Isabel", "Castro", "David", "Edison", "Andres"]
print (lista_de_nombres)
```

```python
print (lista_de_nombres[3])
```

```python
#imprimir el ultimo elemento de la lista
print (lista_de_nombres[4])
```

```python
print (lista_de_nombres)
print (lista_de_nombres[-3])
```

```python
# agregar elementos a lista (append)
lista_de_nombres.append("Carlos")
print(lista_de_nombres)
```

```python
#Reemplazar un elemento
print (lista_de_nombres)
lista_de_nombres [3] = "Juan"
print(lista_de_nombres)
```

```python
# Insertar un elemento
print (lista_de_nombres)
lista_de_nombres.insert(3, "Daniela")
print(lista_de_nombres)
```
